# Pump it Up: Data Mining the Water Table

The goal is to predict the operating condition of a waterpoint for each record in the dataset. You are provided the following set of information about the waterpoints:

- amount_tsh - Total static head (amount water available to waterpoint)
- date_recorded - The date the row was entered
- funder - Who funded the well
- gps_height - Altitude of the well
- installer - Organization that installed the well
- longitude - GPS coordinate
- latitude - GPS coordinate
- wpt_name - Name of the waterpoint if there is one
- num_private -
- basin - Geographic water basin
- subvillage - Geographic location
- region - Geographic location
- region_code - Geographic location (coded)
- district_code - Geographic location (coded)
- lga - Geographic location
- ward - Geographic location
- population - Population around the well
- public_meeting - True/False
- recorded_by - Group entering this row of data
- scheme_management - Who operates the waterpoint
- scheme_name - Who operates the waterpoint
- permit - If the waterpoint is permitted
- construction_year - Year the waterpoint was constructed
- extraction_type - The kind of extraction the waterpoint uses
- extraction_type_group - The kind of extraction the waterpoint uses
- extraction_type_class - The kind of extraction the waterpoint uses
- management - How the waterpoint is managed
- management_group - How the waterpoint is managed
- payment - What the water costs
- payment_type - What the water costs
- water_quality - The quality of the water
- quality_group - The quality of the water
- quantity - The quantity of water
- quantity_group - The quantity of water
- source - The source of the water
- source_type - The source of the water
- source_class - The source of the water
- waterpoint_type - The kind of waterpoint
- waterpoint_type_group - The kind of waterpoint

# Explore the target
The labels in this dataset are simple. There are three possible values:

- `functional` - the waterpoint is operational and there are no repairs needed

## Feature Engineering

- latitud and longitude: have allowed us to find possible duplicated data. However, as it is, these features seems not usefull at all. But before dropping them, I will analyse it through a scatter plot, and see if there could be some kind of usefull information.

First, make a plot trying to see if there are regional patterns.....

```python
df_train['longitude'][df_train['longitude']==0]=np.nan
df_train['latitude'][df_train['latitude']>-2e-05]=np.nan

plt.figure(figsize=(14,10))
sns.scatterplot(x=df_train.longitude, y=df_train.latitude, hue=df_train.status_group, alpha=0.3)
plt.show()
```

Ok....so there are clar regional patterns, which I am not completely sure that correpsonds with any geographyc area (region, district, whatever...). So, I will apply k-means clusttering to obtain different groups, and then we can see if that groups shows different target patterns....

First of all, we fill the NaNs (lat and lon with 0!) using the mean value of the highest level of granularility of the geographic features....

```python
# impute Nans funder that installed teh well: funder > installer > mean rest
df_train[['latitude','longitude']] = df_train[['latitude','longitude']].fillna(df_train.groupby('subvillage')[['latitude','longitude']].transform('mean'))
df_test[['latitude','longitude']] = df_test[['latitude','longitude']].fillna(df_test.groupby('subvillage')[['latitude','longitude']].transform('mean'))
print ('train lon nans...',df_train.longitude.isnull().sum(), '-- train lat nans...',df_train.latitude.isnull().sum())
print ('test lon nans...',df_test.longitude.isnull().sum(), '-- test lat nans...',df_test.latitude.isnull().sum())

df_train[['latitude','longitude']] = df_train[['latitude','longitude']].fillna(df_train.groupby('ward')[['latitude','longitude']].transform('mean'))
df_test[['latitude','longitude']] = df_test[['latitude','longitude']].fillna(df_test.groupby('ward')[['latitude','longitude']].transform('mean'))
print ('train lon nans...',df_train.longitude.isnull().sum(), '-- train lat nans...',df_train.latitude.isnull().sum())
print ('test lon nans...',df_test.longitude.isnull().sum(), '-- test lat nans...',df_test.latitude.isnull().sum())

df_train[['latitude','longitude']] = df_train[['latitude','longitude']].fillna(df_train.groupby('lga')[['latitude','longitude']].transform('mean'))
df_test[['latitude','longitude']] = df_test[['latitude','longitude']].fillna(df_test.groupby('lga')[['latitude','longitude']].transform('mean'))
print ('train lon nans...',df_train.longitude.isnull().sum(), '-- train lat nans...',df_train.latitude.isnull().sum())
print ('test lon nans...',df_test.longitude.isnull().sum(), '-- test lat nans...',df_test.latitude.isnull().sum())
```

So, as it is, this geographic coordinates are not usefull to any ML model. But, we can make clusters and then check if the clustters are able to repdocuce these geographic variability...

```python
kmeans = KMeans(n_clusters=12, random_state=0).fit(df_train[['latitude','longitude']])
df_train['COORD_kmeans']=kmeans.labels_
df_test['COORD_kmeans']=kmeans.predict(df_test[['latitude','longitude']])

# define them as object type
df_train.COORD_kmeans = df_train.COORD_kmeans.astype('object')
df_test.COORD_kmeans = df_test.COORD_kmeans.astype('object')

# some plot to see the groups
plt.figure(figsize=(7,5))
sns.scatterplot(x=df_train.longitude, y=df_train.latitude, hue=df_train.COORD_kmeans, alpha=0.3,
        palette=sns.color_palette("bright",12))
# Put the legend out of the figure
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.show()

plt.figure(figsize=(7,5))
sns.scatterplot(x=df_train.longitude, y=df_train.latitude, hue=df_train.status_group, alpha=0.3)
plt.show()
```

<font color='red'> **drop data 2/5: `latitude`, `longitude`** </font>

```python
# FUNCTION TO PLOT BOX-PLOT o SCATTERPLOT
def plot_boxplot(n,m,df,figx,figy,y_val,box=1):
    """This function plot a multple subplot (defined by n & m) for a dataframe, also defining the size of the figure.
    The plot could be boxplot (box=1) for categorical data, and regplot for numerical data"""
    fig, axs = plt.subplots(n,m,figsize=(figx,figy))
    rect = fig.patch
    rect.set_facecolor('white')
    axs = axs.flatten()
    
    ll= df.columns.to_list()
    ll.remove(y_val)

    for i,(ax, feature) in enumerate (zip(axs, ll)):
        boxplot = sns.countplot(data=df, x=feature, hue='status_group',ax=ax)
                    
        boxplot.set_xticklabels(boxplot.get_xticklabels(), rotation=45, horizontalalignment='right')
        #boxplot.legend()
    plt.tight_layout()
```

```python
# FUNCTION TO PLOT THE CONFUSION MATRIX
def plot_confusion_matrix(cm, classes,
                          normalize=False,
                          title='Confusion matrix',
                          cmap=plt.cm.Oranges):
    """
    This function prints and plots the confusion matrix.
    Normalization can be applied by setting `normalize=True`.
    Source: http://scikit-learn.org/stable/auto_examples/model_selection/plot_confusion_matrix.html
    """
    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
        print("Normalized confusion matrix")
    else:
        print('Confusion matrix, without normalization')

    plt.figure(figsize = (6, 6))
    plt.imshow(cm, interpolation='nearest', cmap='Greens')
    plt.title(title, size = 12)
    plt.colorbar(aspect=4)
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45, size = 12)
    plt.yticks(tick_marks, classes, size = 14)

    fmt = '.2f' if normalize else 'd'
    thresh = cm.max() / 2.
    
    # Labeling the plot
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, format(cm[i, j], fmt), fontsize = 12,
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")
        
    plt.grid(None)
    plt.tight_layout()
    plt.ylabel('True label', size = 12)
    plt.xlabel('Predicted label', size = 12)
```

## Feature Selection

One of the good things of RandomForest (or similar models based on DecisionTrees) is that it allows us to describe wich are the most relevant features (the features that affects the results).

Once we know this, we can reduce our dataset and only use some of the features. This probably would help de model to avoid overfitting, reduce computational time and probably avoid other kind of problems (for example: multicolinealirity)

```python
rdf
```

```python
importances = rdf.feature_importances_
pd.Series(rdf.feature_importances_, index=X_train.columns).nlargest(60).plot(figsize=(10,15),kind='barh')
plt.gca().invert_yaxis()
plt.show()
```

This plot shows us the fetaure importance (the first 300 hundred features), from more to less importance.

Now, we can use a feature selector. In this case, the SelectFromModel from Scikit learn uses the model that we have applied to select only the features that has an effect higher than the one defined in the threshold. 

```python
# use class SelectFromModel. Here the important parameter is the threshold!
sfm = SelectFromModel(rdf, threshold=0.0003)

# Train the selector
sfm.fit(X_train, y_train)
```

```python
# Print the names of the most important features
feat_labels = X_train.columns
new_label=[]
for feature_list_index in sfm.get_support(indices=True):
    print(feat_labels[feature_list_index])
    new_label.append(feat_labels[feature_list_index])
```

```python
X_important_train = sfm.transform(X_train)
X_important_test = sfm.transform(X_test)

X_important_train.shape
```

Applying the threshold, the size of the data has been reduced to 233 features! I hope that with this reduction the model perform well and quicker! (some tests that I've done with less features did not improve the results)

```python
# Create a new random forest classifier for the most important features
rdf_important = RandomForestClassifier(n_estimators=400,criterion='entropy',random_state=42,n_jobs=-1)

#CV R2 
results_accuracy = cross_val_score(rdf_important, X_important_train, y_train, cv=5, 
                                   scoring='accuracy',verbose=1)

print (' ########      CROSS VALIDATION      ############## ')
print('R^2 (coefficient of determination) regression score function for the cross validation:',results_accuracy.mean())

# FIT AND PREDICT
rdf_important.fit(X_important_train,y_train)
pred_rdf_important= rdf_important.predict(X_important_test)
```

```python
plot_confusion_matrix(confusion_matrix(y_test, pred_rdf_important), 
                      classes=[0,1,2],title = 'RandomForest', normalize=False)
plt.show()
```

The results with reduced dataframe is very similar with the entire dataframe! So we can use this with the XGBoost model
